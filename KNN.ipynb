# Data Loading 

```python
import pandas as pd
```

```python
df = pd.read_excel("‏iris.xlsx")
```

# Taking train_set and test_set

```python
from sklearn.model_selection import train_test_split
```

```python
train_set,test_set = train_test_split(df,test_size=0.2,random_state=42)
```

```python
train_set
```

```python
test_set
```

# Taking feature and labels from train_set

```python
feature = train_set.drop('class',axis=1)
```

```python
labels = train_set['class']
```

```python
feature
```

```python
labels
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_feature = scaler.fit_transform(feature)
```

```python
scaled_feature
```

# Train and Evaluate KNN for Diffrent K

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
k_values = [1,3,5,7,9]
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(scaled_feature,labels)
    prediction = knn.predict(scaled_feature)
    print(f"\nk = {k}")
    print("Accuracy:",accuracy_score(labels,prediction))
    cm = confusion_matrix(labels,prediction)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels= labels)
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Set the style for better aesthetics
plt.style.use('ggplot')
sns.set(style="whitegrid")

# Create a figure with multiple subplots
fig = plt.figure(figsize=(15, 12))

# 1. Line Plot
ax1 = fig.add_subplot(2, 2, 1)
x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x), label='sin(x)')
ax1.plot(x, np.cos(x), label='cos(x)')
ax1.set_title('Line Plot')
ax1.legend()

# 2. Bar Chart
ax2 = fig.add_subplot(2, 2, 2)
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 55, 15]
ax2.bar(categories, values, color='skyblue')
ax2.set_title('Bar Chart')

# 3. Scatter Plot
ax3 = fig.add_subplot(2, 2, 3)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
ax3.scatter(x, y, c=colors, s=sizes, alpha=0.7)
ax3.set_title('Scatter Plot')

# 4. Pie Chart
ax4 = fig.add_subplot(2, 2, 4)
labels = ['Group A', 'Group B', 'Group C', 'Group D']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # explode the 2nd slice
ax4.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax4.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
ax4.set_title('Pie Chart')

plt.tight_layout()
plt.show()

# Bonus: Seaborn visualization
plt.figure(figsize=(10, 6))
# Create sample data
tips = sns.load_dataset("tips")
# Create a more complex visualization
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.title('Box Plot with Seaborn')
plt.show()
```
